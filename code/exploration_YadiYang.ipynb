How many transactions are in GBP?

```python
#curr is a table counting transactions in each currency
curr = transactions.groupby(['CURRENCY']).size().reset_index()

#reset the column name
curr.columns = ['CURRENCY','number of transactions']

#get row of gbp currency
gbp = curr.loc[curr['CURRENCY'] == 'GBP']

#get the count of that row
t = gbp.iloc[0]['number of transactions']
print("There are "+str(t)+" transactions in GBP.")
```

How many transactions are NOT in USD?

```python
#total transactions in all currencies
summ = curr['number of transactions'].sum()

#total transactions in usd
usd = curr.loc[curr['CURRENCY'] == 'USD']
us = usd.iloc[0]['number of transactions']

#total transactions not in usd
t = summ-us
print("There are "+str(t)+" transactions not in USD.")
```

What is the average and mediam transaction in USD?

```python
#average and mediam transaction in USD
avg = transactions['AMOUNT_USD'].mean()
median = transactions['AMOUNT_USD'].median()
print("Average of transaction in USD is "+str(round(avg,3)))
print("Median of transaction in USD is "+str(median))
```
