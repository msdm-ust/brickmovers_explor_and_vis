```python
german_credit_data = pd.read_csv('/Users/jiayi/Downloads/german_credit_data.csv')
german_credit_data
```

### What are mean age, mean credit amount, and duration?

```python
age_avg = german_credit_data['Age'].mean()
credit_amount_avg = german_credit_data['Credit amount'].mean()
duration_avg = german_credit_data['Duration'].mean()
print('mean age is',age_avg)
print('mean credit amount is',credit_amount_avg)
print('mean duration is',duration_avg)
```

### What are the major three purpose of loan?

```python
purpose_loan = german_credit_data.groupby(['Purpose'])['Unnamed: 0'].agg([('purpose_cnt','count')]).reset_index()
purpose_loan.sort_values('purpose_cnt',ascending = False)[0:3]
```

### What is the majoriry loan taker? Male of female?

```python
sex_loan = german_credit_data.groupby(['Sex'])['Unnamed: 0'].agg([('sex_cnt','count')]).reset_index()
sex_loan.sort_values('sex_cnt',ascending = False)[0:3]
```

###### the majority of loan taker is male.
